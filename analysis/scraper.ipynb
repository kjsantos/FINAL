```python
import pandas as pd
import selenium 
from selenium import webdriver
import numpy as np
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
driver.get("https://www.opensecrets.org/industries/summary.php?ind=Q&cycle=2022&recipdetail=M&sortorder=U")

codes = []
industries = []

results = driver.find_elements_by_xpath('//*[@id="indus_switch"]')
for r in results:
    opt = r.find_elements_by_tag_name("option")
    for x in opt:
        codes.append(x.get_attribute("value")[:-2])
        industries.append(x.text)
```

```python
codes = ['Q14', 'Q15', 'Q16', 'Z1200', 'J2100', 'Q02', 'Q11', 'Q04', 'Q12', 'Q13', 'Q09', 'Q03', 'J7300', 'Q05', 'Z1100', 'J2200', 'Q01', 'Q08']
industries = ['Abortion Policy/Anti-Abortion', 'Abortion Policy/Pro-Abortion Rights', 'Candidate Committees', 'Democratic Candidate Committees', 'Democratic leadership PAC', 'Democratic/Liberal', 'Environment', 'Foreign & Defense Policy', 'Gun Control', 'Gun Rights', 'Human Rights', 'Leadership PACs', 'LGBTQIA Rights & Issues', 'Pro-Israel', 'Republican Candidate Committees', 'Republican leadership PAC', 'Republican/Conservative', "Women's Issues"]
print(codes, industries)
```

```python
links = ["https://www.opensecrets.org/industries/summary.php?ind=Q&cycle=2022&recipdetail=S&mem=N",
"https://www.opensecrets.org/industries/summary.php?ind=Q&cycle=2022&recipdetail=S&sortorder=A&mem=N&page=2",
"https://www.opensecrets.org/industries/summary.php?ind=Q&cycle=2022&recipdetail=S&sortorder=A&mem=N&page=3"]

sen_cd = []

for link in links:
    driver.get(link)


    results = driver.find_elements_by_xpath('//*[@id="profileLeftColumn"]/table')
    for r in results:
        opt = r.find_elements_by_tag_name("td")
        for i in range (len(opt)):
            if ((i%2==0) or (i%2==2)):
                sen_cd.append(opt[i].text)

sen_cd.sort()
print(sen_cd)
```

```python
candidate_df = pd.DataFrame(columns=industries)
```

```python
candidate_df.insert(0, "Candidates", sen_cd)
```

```python
def scrape(code, page, industry):
    
    
    
    link = "https://www.opensecrets.org/industries/summary.php?ind=" + code + "&cycle=2022&recipdetail=S&sortorder=N&mem=N&page=" + str(page)
    print(link)
    driver.get(link)
    
    totals = []
    
    driver.implicitly_wait(2)
    results = driver.find_elements_by_xpath('//*[@id="profileLeftColumn"]/table')
    for r in results:
        opt = r.find_elements_by_tag_name("td")
        for i in range (len(opt)):
            if (i%2==0):
                print(opt[i])
                ind = np.where(candidate_df["Candidates"] == opt[i].text)
                print(ind)
                amt = (opt[i+1].text[1:]).replace(",", "")
                print(amt)
                candidate_df.at[candidate_df.index[ind[0]], industry] = int(amt)
```

```python
for i in range (len(codes)):
    code = codes[i]
    industry = industries[i]
        
    
    for n in range(1,4):
        scrape(code, n, industry)
        
    #candidate_df[industry] = totals
```

```python
h_link = "https://www.opensecrets.org/industries/summary.php?ind=Q03&cycle=2022&recipdetail=H&sortorder=N&mem=N&page="

house_cd = []

for i in range(1,9):
    driver.get(h_link + str(i))


    results = driver.find_elements_by_xpath('//*[@id="profileLeftColumn"]/table')
    for r in results:
        opt = r.find_elements_by_tag_name("td")
        for i in range (len(opt)):
            if ((i%2==0) or (i%2==2)):
                house_cd.append(opt[i].text)

house_cd.sort()
print(house_cd)
```

```python
house_df = pd.DataFrame(columns=industries)
house_df.insert(0, "Candidates", house_cd)
```

```python
def h_scrape(code, page, industry):
    
    
    
    link = "https://www.opensecrets.org/industries/summary.php?ind=" + code + "&cycle=2022&recipdetail=H&sortorder=N&mem=N&page=" + str(page)
    print(link)
    driver.get(link)
    
    totals = []
    
    driver.implicitly_wait(1)
    results = driver.find_elements_by_xpath('//*[@id="profileLeftColumn"]/table')
    for r in results:
        if (r.text == ""):            
            break
        else:
            opt = r.find_elements_by_tag_name("td")
            for i in range (len(opt)):
                if (i%2==0):
                    ind = np.where(house_df["Candidates"] == opt[i].text)
                    amt = (opt[i+1].text[1:]).replace(",", "")
                    print(amt)
                    house_df.at[house_df.index[ind[0]], industry] = int(amt)

h_scrape("Q14", 10, "Abortion Policy/Anti-Abortion")
```

```python
for i in range (len(codes)):
    code = codes[i]
    industry = industries[i]
        
    
    for n in range(1,9):
        h_scrape(code, n, industry)
```
