```python
link = "https://www.opensecrets.org/members-of-congress/members-list?cong_no=117&cycle=2022&t0-search="

sen_df = pd.read_json('senators.json')
house_df = pd.read_csv('house.csv')
```

```python
def search_cd():

    lname = sen_df['lname'].tolist()
    fname = sen_df['fname'].tolist()
    result = []
    for i in range(len(fname)):
        driver.get(link + (fname[i] + lname[i]))
        driver.implicitly_wait(1)
        results = driver.find_elements_by_xpath('//*[@id="DataTables_Table_0"]/tbody/tr/td[1]')
        for r in results:
            if (r.text == "No matching records found"):
                result.append(0)
            else:
                result.append(1)
    return result
                
    
swl = search_cd()
```

```python
len(swl)
```

```python
sen_df.insert(22, "result", swl)
```
